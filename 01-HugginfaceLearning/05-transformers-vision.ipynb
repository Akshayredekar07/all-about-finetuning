```python
from transformers import pipeline
from PIL import Image
import requests
from io import BytesIO

def load_image(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers, timeout=10)
    response.raise_for_status()
    return Image.open(BytesIO(response.content)).convert("RGB")

# ── 1. google/vit-base-patch16-224 ───────────────────────────────────────────
pipe = pipeline("image-classification", model="google/vit-base-patch16-224")

urls = {
    "dog"      : "https://upload.wikimedia.org/wikipedia/commons/2/26/YellowLabradorLooking_new.jpg",
    "cat"      : "https://upload.wikimedia.org/wikipedia/commons/4/4d/Cat_November_2010-1a.jpg",
    "sunflower": "https://upload.wikimedia.org/wikipedia/commons/4/41/Sunflower_from_Silesia2.jpg",
}

print("── Classification from URL (with headers) ──")
for name, url in urls.items():
    img = load_image(url)
    results = pipe(img, top_k=3)
    print(f"Image  : {name}")
    for r in results:
        print(f"  [{r['score']:.4f}]  {r['label']}")
    print()
```

```python
# from local file path
print("── Classification from file path ──")
result = pipe("./kingfisher.png", top_k=3)
for r in result:
    print(f"  [{r['score']:.4f}]  {r['label']}")
print()
```

```python
pipe = pipeline("image-classification", model="microsoft/resnet-50")

urls = {
    "dog"      : "https://upload.wikimedia.org/wikipedia/commons/2/26/YellowLabradorLooking_new.jpg",
    "ant"      : "https://upload.wikimedia.org/wikipedia/commons/a/a7/Camponotus_flavomarginatus_ant.jpg",
    "sunflower": "https://upload.wikimedia.org/wikipedia/commons/4/41/Sunflower_from_Silesia2.jpg",
}

print("── ResNet-50 predictions ──")
for name, url in urls.items():
    img = load_image(url)
    results = pipe(img, top_k=3)
    print(f"Image  : {name}")
    for r in results:
        print(f"  [{r['score']:.4f}]  {r['label']}")
    print()
```

```python
# ── 3. facebook/convnext-base-224 ─────────────────────────────────────────────
#    Modern pure-CNN — ConvNeXt design inspired by ViT, no attention needed
#    Best for: strong CNN baseline, better than ResNet on most benchmarks
# ─────────────────────────────────────────────────────────────────────────────
pipe = pipeline("image-classification", model="facebook/convnext-base-224")

urls = {
    "cat"     : "https://upload.wikimedia.org/wikipedia/commons/1/14/Gatto_europeo4.jpg",
    # "rabbit"  : "https://upload.wikimedia.org/wikipedia/commons/b/bf/Domestic_rabbit.jpg",
    "sunflower": "https://upload.wikimedia.org/wikipedia/commons/4/41/Sunflower_from_Silesia2.jpg",
}

print("── ConvNeXt predictions ──")
for name, url in urls.items():
    img = load_image(url)
    results = pipe(img, top_k=3)
    print(f"Image  : {name}")
    for r in results:
        print(f"  [{r['score']:.4f}]  {r['label']}")
    print()

```

```python
# ── 4. apple/mobilevit-small ──────────────────────────────────────────────────
#    Hybrid CNN + ViT — lightweight, designed for mobile and edge deployment
#    Best for: resource-constrained environments, fast CPU inference
# ─────────────────────────────────────────────────────────────────────────────
pipe = pipeline("image-classification", model="apple/mobilevit-small")

urls = {
    "dog": "https://upload.wikimedia.org/wikipedia/commons/2/26/YellowLabradorLooking_new.jpg",
    "cat": "https://upload.wikimedia.org/wikipedia/commons/4/4d/Cat_November_2010-1a.jpg",
    "ant": "https://upload.wikimedia.org/wikipedia/commons/a/a7/Camponotus_flavomarginatus_ant.jpg",
}

print("── MobileViT (lightweight) predictions ──")
for name, url in urls.items():
    img = load_image(url)
    results = pipe(img, top_k=3)
    print(f"Image  : {name}")
    for r in results:
        print(f"  [{r['score']:.4f}]  {r['label']}")
    print()
```

```python
# ── 5. batch processing ───────────────────────────────────────────────────────
#    Pass a list of PIL Images — processed together, more efficient
#    Best for: classifying large image datasets, production pipelines
# ─────────────────────────────────────────────────────────────────────────────
pipe = pipeline("image-classification", model="google/vit-base-patch16-224", use_fast=True)

named_images = {
    "dog"      : "https://upload.wikimedia.org/wikipedia/commons/2/26/YellowLabradorLooking_new.jpg",
    "cat"      : "https://upload.wikimedia.org/wikipedia/commons/4/4d/Cat_November_2010-1a.jpg",
    "sunflower": "https://upload.wikimedia.org/wikipedia/commons/4/41/Sunflower_from_Silesia2.jpg",
    "ant"      : "https://upload.wikimedia.org/wikipedia/commons/a/a7/Camponotus_flavomarginatus_ant.jpg",
}

images = [load_image(url) for url in named_images.values()]

print("── Batch classification ──")
results = pipe(images, top_k=2)

for name, preds in zip(named_images.keys(), results):
    top = preds[0]
    print(f"[{top['score']:.4f}] {top['label']:30s}  ←  {name}")
print()
```
