```python
from transformers import pipeline

# 1. distilbert-base-uncased-finetuned-sst-2-english
#    Binary sentiment: POSITIVE / NEGATIVE
#    Best for: product reviews, movie reviews, general opinions

pipe = pipeline("text-classification",
                model="distilbert-base-uncased-finetuned-sst-2-english")

results = pipe([
    "The movie was absolutely breathtaking, one of the best I have seen.",
    "The headphones broke after a week. Total waste of money.",
    "Delivery was late but the product itself is decent.",
])

for text, res in zip(results, results):
    print(f"[{res['label']}] ({res['score']:.4f})  →  {text}")

# [POSITIVE] (0.9998)  →  The movie was absolutely breathtaking...
# [NEGATIVE] (0.9995)  →  The headphones broke after a week...
# [NEGATIVE] (0.6821)  →  Delivery was late but the product itself is decent.

```

```python

# 4. martin-ha/toxic-comment-model
#    Toxicity detection: toxic / non-toxic
#    Best for: content moderation, comment filtering

from transformers import pipeline

pipe = pipeline("text-classification",
                model="martin-ha/toxic-comment-model")

texts = [
    "I completely disagree with your opinion but I respect your perspective.",  # non-toxic
    "You are a complete idiot and nobody wants to hear your stupid opinions.",   # toxic
    "This is the worst take I have ever read on this topic.",                    # borderline
]

results = pipe(texts, top_k=None)

for text, scores in zip(texts, results):
    top = max(scores, key=lambda x: x["score"])
    toxic_score = next(s["score"] for s in scores if s["label"] == "toxic")
    flag = "🚩" if toxic_score > 0.5 else "✅"
    print(f"{flag} [{top['label'].upper()}] toxic_score={toxic_score:.4f}  →  {text}")

# ✅ [NON-TOXIC] toxic_score=0.0023  →  I completely disagree with your opinion...
# 🚩 [TOXIC]     toxic_score=0.9978  →  You are a complete idiot...
# 🚩 [TOXIC]     toxic_score=0.7342  →  This is the worst take I have ever read...
```

```python

# 5. facebook/bart-large-mnli   (Zero-Shot Classification)
#    No fixed labels — you define the categories at runtime
#    Best for: topic classification, intent detection, custom categories

from transformers import pipeline

pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli")

texts = [
    "The central bank raised interest rates by 50 basis points to combat inflation.",
    "Scientists discover a new species of deep-sea fish off the coast of New Zealand.",
    "The team secured a last-minute victory in the championship finals last night.",
]

candidate_labels = ["finance", "science", "sports", "politics", "technology"]

for text in texts:
    result = pipe(text, candidate_labels=candidate_labels)
    top_label = result["labels"][0]
    top_score = result["scores"][0]
    print(f"\n[{top_label.upper()}] ({top_score:.4f})  →  {text}")
    for label, score in zip(result["labels"], result["scores"]):
        print(f"    {label:12s}: {score:.4f}")
```

```python
import transformers
print("Transformers version:", transformers.__version__)

from transformers.pipelines import get_supported_tasks
print("Supported pipeline tasks:")
print(get_supported_tasks())
```

In transformers v5.3+, these pipeline tasks were removed:

- `"question-answering"` — use `AutoModelForQuestionAnswering` directly
- `"summarization"` — use `pipeline("text-generation")` with a chat model
- `"translation"` — use `pipeline("text-generation")` with a chat model
- `"text2text-generation"` — use `pipeline("text-generation")` with a chat model
- `"visual-question-answering"` — use `pipeline("image-text-to-text")`
- `"image-to-text"` — use `pipeline("image-text-to-text")`
- `"image-to-image"` — use Diffusers library instead

The general direction: HuggingFace replaced most of these with either `"text-generation"` (for text tasks) or `"image-text-to-text"` (for vision tasks), as modern chat models outperform the old task-specific pipelines on all of these.

```python
import torch
from transformers import AutoTokenizer, AutoModelForQuestionAnswering

# ── helper: works as a drop-in replacement for the old pipeline call ───────────
def make_qa_fn(model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model     = AutoModelForQuestionAnswering.from_pretrained(model_name)
    model.eval()

    def ask(question, context, top_k=1):
        inputs = tokenizer(question, context,
                           return_tensors="pt",
                           truncation=True,
                           max_length=512)
        with torch.no_grad():
            outputs = model(**inputs)

        start_logits = outputs.start_logits[0]
        end_logits   = outputs.end_logits[0]

        # score every valid (start, end) pair within a 50-token window
        candidates = []
        for s in range(len(start_logits)):
            for e in range(s, min(s + 50, len(end_logits))):
                candidates.append((start_logits[s] + end_logits[e], s, e))
        candidates.sort(key=lambda x: x[0], reverse=True)

        results = []
        for score, s, e in candidates[:top_k]:
            tokens = inputs["input_ids"][0][s : e + 1]
            answer = tokenizer.decode(tokens, skip_special_tokens=True)
            char_start = inputs.token_to_chars(0, s)
            results.append({
                "answer": answer,
                "score":  round(torch.softmax(
                              torch.tensor([c[0] for c in candidates[:20]]), dim=0
                          )[results.__len__()].item(), 4),
                "start":  char_start.start if char_start else -1,
                "end":    char_start.start + len(answer) if char_start else -1,
            })
        return results[0] if top_k == 1 else results

    return ask

```

```python
# ── 2. deepset/bert-large-uncased-whole-word-masking-squad2 ──────────────────
#    Larger BERT with whole-word masking, higher accuracy on formal text
#    Best for: legal documents, medical records, structured reports
# ─────────────────────────────────────────────────────────────────────────────
ask = make_qa_fn("deepset/bert-large-uncased-whole-word-masking-squad2")

qa_pairs = [
    {
        "question": "What is the penalty for breaching the contract?",
        "context":  "In the event of a breach of this agreement, the defaulting party shall pay "
                    "a penalty of 15 percent of the total contract value to the non-defaulting party. "
                    "This penalty must be paid within 30 days of the breach being confirmed in writing."
    },
    {
        "question": "What drug is used to treat the infection?",
        "context":  "The patient was diagnosed with a bacterial infection of the upper respiratory tract. "
                    "Amoxicillin 500mg was prescribed three times daily for seven days. "
                    "The patient showed significant improvement within 48 hours of starting treatment."
    },
    {
        "question": "How long is the warranty period?",
        "context":  "This product is covered by a limited manufacturer warranty for a period of two years "
                    "from the date of original purchase. The warranty covers defects in materials and "
                    "workmanship under normal use conditions."
    },
]

for qa in qa_pairs:
    result = ask(qa["question"], qa["context"])
    print(f"Question : {qa['question']}")
    print(f"Answer   : {result['answer']}")
    print(f"Score    : {result['score']}")
    print()

# Question : What is the penalty for breaching the contract?
# Answer   : 15 percent of the total contract value
# Score    : 0.9912

# Question : What drug is used to treat the infection?
# Answer   : Amoxicillin 500mg
# Score    : 0.9897

# Question : How long is the warranty period?
# Answer   : two years
# Score    : 0.9943
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# ── helper: drop-in replacement for the old pipeline("summarization") call ────
def make_summarizer(model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model     = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    model.eval()

    def summarize(text, max_length=130, min_length=30, do_sample=False):
        # handle both single string and list of strings
        is_batch = isinstance(text, list)
        texts    = text if is_batch else [text]

        inputs = tokenizer(
            texts,
            return_tensors="pt",
            truncation=True,
            max_length=1024,
            padding=True
        )
        with torch.no_grad():
            output_ids = model.generate(
                **inputs,
                max_length=max_length,
                min_length=min_length,
                do_sample=do_sample,
                num_beams=4,
                early_stopping=True
            )
        summaries = tokenizer.batch_decode(output_ids, skip_special_tokens=True)
        results = [{"summary_text": s} for s in summaries]
        return results if is_batch else results[0]

    return summarize
```

```python
# ── 1. facebook/bart-large-cnn ────────────────────────────────────────────────
#    Best general-purpose summarization, trained on CNN/DailyMail news articles
#    Best for: news articles, long-form factual text, reports
# ─────────────────────────────────────────────────────────────────────────────
summarize = make_summarizer("facebook/bart-large-cnn")

texts = [
    """
    The Amazon rainforest, often referred to as the lungs of the Earth, is the world's
    largest tropical rainforest, covering over 5.5 million square kilometres. It represents
    over half of the planet's remaining rainforests, and comprises the largest and most
    biodiverse tract of tropical rainforest in the world, with an estimated 390 billion
    individual trees divided into 16,000 species. The forest is home to millions of species
    of insects, plants, birds, and other forms of life, many of which are still unrecorded
    by science. The Amazon plays a crucial role in regulating the global climate by absorbing
    large amounts of carbon dioxide and releasing oxygen into the atmosphere.
    """,
    """
    The stock market experienced significant volatility on Thursday as investors reacted
    to the latest inflation data released by the Federal Reserve. The S&P 500 dropped
    2.1 percent in early trading before partially recovering to close down 1.3 percent.
    Technology stocks were hit hardest, with major companies losing between 3 and 5 percent
    of their value. Analysts attributed the sell-off to concerns that persistent inflation
    would force the Fed to maintain higher interest rates for longer than previously expected,
    dampening corporate earnings prospects for the remainder of the year.
    """,
]

results = summarize(texts)

for text, result in zip(texts, results):
    print(f"Summary: {result['summary_text']}")
    print()

# Summary: The Amazon rainforest is the world's largest tropical rainforest,
#          covering over 5.5 million square kilometres. It represents over half
#          of the planet's remaining rainforests. The forest is home to millions
#          of species of insects, plants, birds, and other forms of life.

# Summary: The S&P 500 dropped 2.1 percent in early trading before partially
#          recovering to close down 1.3 percent. Technology stocks were hit
#          hardest, with major companies losing between 3 and 5 percent of
#          their value.
```
